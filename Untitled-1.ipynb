```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf

# Load the dataset
df = pd.read_excel("t07.csv.xlsx")

# Create a binary variable 'east_west' based on smoker or drinker status
df['east_west'] = np.where((df['Social_smoker'] == 'yes') | (df['Social_drinker'] == 'yes'), 1, 0)

# Convert categorical variables into dummy/indicator variables
categorical_columns = ['Disciplined', 'Social_smoker', 'Social_drinker']
df = pd.get_dummies(df, columns=categorical_columns, drop_first=True, dtype=int)

# Define the formula for OLS regression
# We'll use all columns except 'Months_active' as independent variables
independent_vars = [col for col in df.columns if col != 'Months_active']
formula = 'Months_active ~ ' + ' + '.join(independent_vars)

# Fit the OLS model
model = smf.ols(formula=formula, data=df).fit()

# Print model summary
print(model.summary())
```

```python
import pandas as pd
import numpy as np
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif
from statsmodels.tools.tools import add_constant

# Load the dataset
df = pd.read_excel("t07.csv.xlsx")

# Create a binary variable 'east_west' based on smoker or drinker status
df['east_west'] = np.where((df['Social_smoker'] == 'yes') | (df['Social_drinker'] == 'yes'), 1, 0)

# Convert categorical variables into dummy variables
categorical_columns = ['Disciplined', 'Social_smoker', 'Social_drinker']
df = pd.get_dummies(df, columns=categorical_columns, drop_first=True, dtype=int)

# Select the independent variables used in your regression model
X = df[['Age', 'BMI', 'Children', 'east_west', 'Disciplined_yes',
        'Social_smoker_yes', 'Social_drinker_yes']]

# Add constant term (required for VIF calculation)
X = add_constant(X)

# Create a DataFrame to store VIF results
VIF_data = pd.DataFrame()
VIF_data["variable"] = X.columns
VIF_data["VIF"] = [vif(X.values, i) for i in range(X.shape[1])]

# Print the results
print(VIF_data)
```

```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf

# Load your Excel file
df = pd.read_excel("t07.csv.xlsx")

# Create the derived 'east_west' variable
df['east_west'] = np.where((df['Social_smoker'] == 'yes') | (df['Social_drinker'] == 'yes'), 1, 0)

# Convert categorical variables into dummy variables
df = pd.get_dummies(df, columns=['Disciplined', 'Social_smoker', 'Social_drinker'], drop_first=True, dtype=int)

# Define and fit the multiple regression model
model = smf.ols(formula="""
    Months_active ~ Distance_from_work + Age + Children + Pets + Weight + Height +
    BMI + Absent_hours + east_west + Disciplined_yes + Social_smoker_yes + Social_drinker_yes
""", data=df).fit()

# Create a DataFrame showing coefficients, their absolute values, and p-values
summary_df = pd.DataFrame({
    "Coefficient": model.params,
    "Absolute Coefficient": model.params.abs(),
    "p-value": model.pvalues
})

# Drop intercept and sort by strength of effect
summary_df = summary_df.drop("Intercept").sort_values(by="Absolute Coefficient", ascending=False)

# Display the table
print(summary_df)
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import make_scorer, mean_absolute_error

# Load the dataset
df = pd.read_excel("t07.csv.xlsx")

# Create 'east_west' variable
df['east_west'] = np.where((df['Social_smoker'] == 'yes') | (df['Social_drinker'] == 'yes'), 1, 0)

# Convert categorical columns into dummy variables
df = pd.get_dummies(df, columns=['Disciplined', 'Social_smoker', 'Social_drinker'], drop_first=True, dtype=int)

# Define features and target variable
X = df.drop(columns=['Months_active'])
y = df['Months_active']

# Set up the linear regression model
model = LinearRegression()

# Use 5-fold cross-validation
cv = KFold(n_splits=5, shuffle=True, random_state=42)

# Use mean absolute error as the scoring method
mae_scorer = make_scorer(mean_absolute_error)

# Perform cross-validation
mae_scores = cross_val_score(model, X, y, scoring=mae_scorer, cv=cv)

# Calculate the average prediction error (mean absolute deviation)
average_mae = np.mean(mae_scores)
print(f"Average prediction error (Mean Absolute Deviation): {average_mae:.2f} months")
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV, RepeatedKFold
from sklearn.preprocessing import MinMaxScaler

# Load the dataset
df = pd.read_excel("t07.csv.xlsx")

# Drop missing values
df = df.dropna()

# Create 'east_west' feature
df['east_west'] = np.where((df['Social_smoker'] == 'yes') | 
                           (df['Social_drinker'] == 'yes'), 1, 0)

# Create dummy variables for categorical features
df = pd.get_dummies(df, columns=['Disciplined', 'Social_smoker', 'Social_drinker'], 
                    drop_first=True, dtype=int)

# Define target and features
y = df['Months_active']
X = df.drop(columns=['Months_active'])

# Normalize feature data
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Define Ridge regression and grid search parameters
ridge = Ridge()
param_grid = {'alpha': np.linspace(0.01, 10, 100)}  # Try 100 values between 0.01 and 10
cv = RepeatedKFold(n_splits=5, n_repeats=1, random_state=42)

# Perform grid search with cross-validation
grid = GridSearchCV(estimator=ridge, param_grid=param_grid, 
                    scoring='neg_mean_absolute_error', cv=cv)
grid.fit(X_scaled, y)

# Output the best alpha value
print(f"Best alpha value: {grid.best_params_['alpha']}")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

# Load your dataset
df = pd.read_excel("t07.csv.xlsx")

# Clean data
df = df.dropna()
df['east_west'] = np.where((df['Social_smoker'] == 'yes') | 
                           (df['Social_drinker'] == 'yes'), 1, 0)
df = pd.get_dummies(df, columns=['Disciplined', 'Social_smoker', 'Social_drinker'], drop_first=True, dtype=int)

# Define target and features
y = df['Months_active']  # or replace with 'Selling_Price' if that's your actual target
X = df.drop(columns=['Months_active'])

# Normalize features
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Define neural network with specified hyperparameters
model = Sequential()
model.add(Dense(512, activation='relu', input_shape=(X_train.shape[1],)))

# Add 4 more hidden layers
for _ in range(4):
    model.add(Dense(512, activation='relu'))

# Output layer
model.add(Dense(1))  # One neuron for regression output

# Compile the model
model.compile(optimizer=Adam(), loss='mean_absolute_error')

# Train the model
history = model.fit(X_train, y_train, 
                    validation_data=(X_test, y_test),
                    epochs=100, 
                    batch_size=16,
                    verbose=1)


```

```python
from sklearn.metrics import mean_absolute_error

# Predict the values using the neural network model
y_pred = model.predict(X_test)

# Calculate the Mean Absolute Error
mae = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error (MAE) from the neural network: {mae:.2f} months")
```
